```python
# Librerias necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Datos

```python
# Lectura del dataframe de taquilla
df_data = pd.read_csv('../data/GastosCancunFC_Data.csv')

df_data.head()
```

```python
# Cambiar formatos de columnas
df_data['Fecha'] = pd.to_datetime(df_data['Fecha'], format='%d-%m-%Y')

df_data.head()
```

```python
df_data.info()
```

```python
df_data["Monto"] = df_data["Monto"].str.replace(",", "").astype(float)  # Convertir a float primero


```

```python
df_data.head()
```

```python
df_data.info()
```

```python
df_data["Cantidad"] = pd.to_numeric(df_data["Cantidad"], errors="coerce").astype("Int64")
```

```python
df_data.info()
```

```python
df_data.head()
```

```python
df_data.to_csv("../data/GastosCancunFC_Data_Limpio.csv", index=False, float_format="%.2f")
```
