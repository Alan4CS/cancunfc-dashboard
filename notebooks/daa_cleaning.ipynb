```python
# Librerias necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Gastos

```python
# Lectura del dataframe de gastos
df_gastos = pd.read_csv('../data/GastosCancunFC_Gastos.csv')

df_gastos.head()
```

```python
# Cambiar formatos de columnas
df_gastos['Fecha'] = pd.to_datetime(df_gastos['Fecha'], format='%d-%m-%Y')

df_gastos.head()
```

```python
# Eliminar comas y convertir a float
df_gastos['Costos'] = df_gastos['Costos'].str.replace(',', '', regex=True).astype(float)

# Confirmar el tipo de dato
print(df_gastos['Costos'].dtype)  # Debe mostrar: float64

# Mostrar los primeros valores para verificar
print(df_gastos[['Costos']].head())
```

```python
df_gastos.to_csv("../data/GastosCancunFC_Gastos_Limpio.csv", index=False, float_format="%.2f")
```

# Taquilla

```python
# Lectura del dataframe de taquilla
df_taquilla = pd.read_csv('../data/GastosCancunFC_Taquilla.csv')

df_taquilla.head()
```

```python
# Cambiar formatos de columnas
df_taquilla['Fecha'] = pd.to_datetime(df_taquilla['Fecha'], format='%d-%m-%Y')

df_taquilla.head()
```

```python
# Convertir la columna a num辿rica, dejando NaN donde est辿 vac鱈o
df_taquilla['Boletos Vendidos'] = pd.to_numeric(df_taquilla['Boletos Vendidos'], errors='coerce')

# Convertir solo los valores NO nulos a enteros, manteniendo los NaN intactos
df_taquilla['Boletos Vendidos'] = df_taquilla['Boletos Vendidos'].astype('Int64')
```

```python
print(df_taquilla[['Boletos Vendidos']].head(10))
```

```python
# Eliminar comas y convertir a float
df_taquilla['Ingreso'] = df_taquilla['Ingreso'].str.replace(',', '', regex=True).astype(float)

# Confirmar el tipo de dato
print(df_taquilla['Ingreso'].dtype)  # Debe mostrar: float64

# Mostrar los primeros valores para verificar
print(df_taquilla[['Ingreso']].head())
```
